```python
#max flow 2, http://nbviewer.jupyter.org/github/Pyomo/PyomoGallery/blob/master/diet/DietProblem.ipynb
from pyomo.environ import *

model = AbstractModel()


# Nodes in the network
model.N = Set()
# Network arcs
model.A = model.N*model.N



# Source node
model.s = Set()
model.t = Set()
model.N11 = model.N - model.s - model.t
model.N1 = model.N *(model.N - model.s - model.t)
model.N2 = model.N*model.t
# Flow capacity limits
model.c = Param(model.A)


# The flow over each arc
model.f = Var(model.A, within=NonNegativeReals)
# Maximize the flow into the sink nodes
def total_rule(model):
    return sum(model.f[i,j] for (i, j) in model.N2)
model.total = Objective(rule=total_rule, sense=maximize)
# Enforce an upper limit on the flow across each arc
def limit_rule(model, i, j):
    return model.f[i,j] <= model.c[i, j]
model.limit = Constraint(model.A, rule=limit_rule)

# Enforce flow through each node
def flow_rule(model, j):
    return sum(model.f[i,j]for (i,j) in model.N1 )== sum(model.f[j,i]for (i,j) in model.N1 )
model.flow = Constraint(model.N11, rule=flow_rule)
!cat Exp2.dat
instance = model.create("Exp2.dat")
instance.pprint()
```
