```python
from __future__ import division
from pyomo.environ import *
from pyomo.opt import SolverFactory
model = AbstractModel()
model.N = Param(within=NonNegativeIntegers)

model.I = RangeSet(1,model.N)
model.J = RangeSet(1,model.N)
model.New = model.I*model.J
#model.New2 = model.New[i,j] for (i,j) in model.New if i!=j
model.II = RangeSet(2,model.N)
model.JJ = RangeSet(2,model.N)

model.c = Param(model.I, model.J)
model.x = Var(model.I, model.J, domain= Binary)
model.u = Var(model.II, domain = NonNegativeIntegers)
model.v = Var(model.JJ, domain = NonNegativeIntegers)

def objj1(model):
    return sum(model.c[i,j]*model.x[i,j] for (i,j) in model.New if i!=j)
model.OBJ = Objective(rule = objj1)
def const1(model,j):
    return sum(model.x[i,j] for i in model.I if i!=j) == 1
model.const1 = Constraint(model.J, rule=const1)
def const2(model,i):
    return sum(model.x[i,j] for j in model.J if i!=j) == 1
model.const2 = Constraint(model.J, rule=const2)
def const3(model,i,j):
    if i!=j:
        return model.u[i]-model.v[j]+model.N*model.x[i,j]  <= model.N-1
model.const3 = Constraint(model.II, model.JJ)
!cat Exp5.dat
#!pyomo solve --solver=glpk Untitled8.py Exp5.dat
#!cat results.yml
instance = model.create("Exp5.dat")
instance.pprint()
opt = SolverFactory('glpk')

results = opt.solve(instance, tee=True)
results.write()
instance.solutions.load_from(results)

for v in instance.component_objects(Var, active=True):
    print ("Variable",v)
    varobject = getattr(instance, str(v))
    for index in varobject:
        print ("   ",index, varobject[index].value)

```
